```python
import cv2
import numpy as np

def find_max_width_and_mark(image_path):
    # 画像を読み込む
    original_image = cv2.imread(image_path)
    resized_image = cv2.resize(original_image, (500, 500))  # 必要に応じてリサイズ

    # 画像を2値化する
    gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
    _, binary_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)

    # 画像の輪郭を抽出
    contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # 最大横幅とそのY座標を初期化
    max_width = 0
    max_width_y = 0

    # 各輪郭に対して横幅を計算し、最大横幅を求める
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(resized_image, (x, y), (x + w, y + h), (0, 0, 255), 2)

        if w > max_width:
            max_width = w
            max_width_y = y

    print("最大横幅:", max_width)

    # 最大横幅があるY座標にマークを付ける
    marked_image = resized_image.copy()
    cv2.line(marked_image, (0, max_width_y), (marked_image.shape[1], max_width_y), (0, 255, 0), 2)

    # 画像保存
    cv2.imwrite("a.png",binary_image)
    cv2.imwrite('marked_image_with_max_width.png', marked_image)
    print("画像を保存しました。")

# 画像のパスを指定して処理を実行
image_path = './drill_test2.png'  # 画像のパスを適切に指定してください
find_max_width_and_mark(image_path)
```

```python
import cv2
import numpy as np

def find_max_width(image_path):
    # 画像を読み込む
    original_image = cv2.imread(image_path)
    resized_image = cv2.resize(original_image, (500, 500))  # 必要に応じてリサイズ

    # 画像を2値化する
    gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
    _, binary_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)

    # 最大横幅とそのY座標を初期化
    max_width = 0
    max_width_y = 0

    target_min_x = 0
    target_max_x = 0

    # 1行ずつ黒色の最小と最大のピクセルを探し、横幅を計算する
    for y in range(binary_image.shape[0]):
        black_pixels = np.where(binary_image[y, :] == 0)[0]
        if len(black_pixels) > 0:
            min_x = np.min(black_pixels)
            max_x = np.max(black_pixels)
            width = max_x - min_x
            if width > max_width:
                target_max_x  = max_x
                target_min_x  = min_x
                max_width = width
                max_width_y = y

    print("最大横幅:", max_width)

    # 最大横幅があるY座標にマークを付ける
    marked_image = resized_image.copy()
    cv2.line(marked_image, (target_min_x, max_width_y), (target_max_x, max_width_y), (0, 255, 0), 2)

    # 画像保存
    cv2.imwrite('marked_image_with_max_width.png', marked_image)
    print("画像を保存しました。")

# 画像のパスを指定して処理を実行
image_path = './imgdrill.png'  # 画像のパスを適切に指定してください
find_max_width(image_path)
```

```python
import cv2
import numpy as np


def template_matching(template_path, target_path, threshold=0.8):
    # テンプレート画像の読み込み
    template = cv2.imread(template_path, 0)

    # 対象画像の読み込み
    target = cv2.imread(target_path, 0)

    # テンプレートマッチングの実行
    result = cv2.matchTemplate(target, template, cv2.TM_CCOEFF_NORMED)

    # 最大値とその位置を取得
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

    # マッチング率が閾値以上の場合に判定
    if max_val >= threshold:
        # テンプレートが最もマッチした位置に矩形を描画
        h, w = template.shape
        top_left = max_loc
        bottom_right = (top_left[0] + w, top_left[1] + h)
        cv2.rectangle(target, top_left, bottom_right, 255, 2)

        # 結果の表示
        cv2.imwrite('Result_templatematching.png', target)
        print("テンプレートが見つかりました。")
    else:
        print("テンプレートは見つかりませんでした。")


if __name__ == "__main__":
    template_path = './imgdrilltemplate.png'
    target_path = './drill_test.png'

    template_matching(template_path, target_path)
```
