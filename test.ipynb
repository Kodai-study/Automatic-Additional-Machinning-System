```python
import cv2
import tkinter as tk
from PIL import Image, ImageTk

# 画像の読み込み
image = cv2.imread("Z:\yubi.JPG")
height, width = image.shape[:2]

# ウィンドウを作成
root = tk.Tk()
root.title("画像ウィンドウ")

# ウィンドウの初期サイズを設定
initial_width = width // 2
initial_height = height // 2
root.geometry(f"{initial_width}x{initial_height}")

# OpenCVの画像をPillowのImageに変換
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_pil = Image.fromarray(image_rgb)

# PillowのImageをTkinter用のImageに変換
photo = ImageTk.PhotoImage(image=image_pil)

# ラベルに画像を表示
label = tk.Label(root, image=photo)
label.pack()

# ウィンドウを表示
root.mainloop()
```

```python
import cv2

# 画像の読み込み
image = cv2.imread("Z:\yubi.JPG")

# 画像を半分のサイズにリサイズ
resized_image = cv2.resize(image, (500, 1000))

# グレースケールに変換
gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

# 画像の輪郭を抽出
contours, _ = cv2.findContours(
    gray_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 輪郭が描かれた新しい画像を作成
contour_image = resized_image.copy()
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)

# 新しい画像の表示
cv2.imshow("Contour Image", contour_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# 画像の読み込み
image = cv2.imread("Z:\drill.png")

# 画像を半分のサイズにリサイズ
resized_image = cv2.resize(image, (500, 1000))

# グレースケールに変換
gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

# 2値化処理を行い、白い背景と黒い指にする
_, binary_image = cv2.threshold(gray_image, 150, 255, cv2.THRESH_BINARY_INV)

# 画像の輪郭を抽出
contours, _ = cv2.findContours(
    binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 輪郭が描かれた新しい画像を作成
contour_image = resized_image.copy()
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)

# 輪郭の直径を計算して表示
for contour in contours:
    (x, y), radius = cv2.minEnclosingCircle(contour)
    diameter = 2 * radius
    print(f"輪郭の直径: {diameter:.2f} ピクセル")
    
    # 輪郭上に直径を描写
    cv2.putText(contour_image, f"{diameter:.2f} ピクセル", (int(x), int(y) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)

# 新しい画像の表示
cv2.imshow("Contour Image", contour_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# 画像の読み込み
image = cv2.imread("Z:\drill.png")

# 画像を半分のサイズにリサイズ
resized_image = cv2.resize(image, (400, 800))

# グレースケールに変換
gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

# 2値化処理を行い、白い背景と黒い指にする
_, binary_image = cv2.threshold(gray_image, 150, 255, cv2.THRESH_BINARY_INV)

# 画像の輪郭を抽出
contours, _ = cv2.findContours(
    binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 輪郭が描かれた新しい画像を作成
contour_image = resized_image.copy()
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)

# 輪郭の直径と横幅を計算して表示
for contour in contours:
    (x, y), radius = cv2.minEnclosingCircle(contour)
    diameter = 2 * radius
    x, y, w, h = cv2.boundingRect(contour)  # 輪郭の外接矩形を取得
    cv2.rectangle(contour_image, (x, y), (x + w, y + h), (0, 0, 255), 2)

print(f"横幅: {w:.2f} ピクセル")
# 新しい画像の表示
cv2.imshow("Contour Image", contour_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
